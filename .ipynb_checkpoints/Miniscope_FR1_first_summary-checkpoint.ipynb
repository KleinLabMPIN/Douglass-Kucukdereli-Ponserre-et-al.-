```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from miniscope.imagingIO import loopMice, loadData, loadBeh, getBeh
from miniscope.analysis import dffCalc, filtData, smoothData, normData
from miniscope.events import getEvents
from miniscope.trials import markTrials
from miniscope.bouts import getBouts, getBoutDur, markBouts
from miniscope.statistics import stdCalc, getBoutMeans, getStats
from miniscope.plotter import plotMean, plotHeat, plotTrials, plotHeatTrials
```

```python
# Start gathering the data from FR1 experiment
# Mouse IDs and dataset IDs
mice = [(8404, 6), (8857, 3), (8864, 1)]
# Parameters
# Behavior type
behType = 'FR1'
# Imaging frame rate
fs = 0.05
# Which trials to include
trials=[1,12]
# Do you want to save?
save = False
```

```python
# Get file paths for imaging data and behavior timestamps
fileList = loopMice(mice, behType)

# Load the imaging data
dataList = loadData(mice, behType)

# Load the behaviour timestamps
eventList = getBeh(mice, fileList['Behaviour'], behType)
```

```python
# Plot the raw data
plt.figure(figsize=(20,10), facecolor="w", dpi= 150)
for i, [mus, sess] in enumerate(mice):
    ax = plt.subplot(len(mice),1,i+1)
    dataList[mus].plot(ax=ax, legend=False, cmap=plt.cm.rainbow)
    ax.set_ylabel("dF/F (%)")
```

```python
# dFF and plot
dFFList = dffCalc(mice, dataList, lowest=False)

# Get the events
eventStart = getEvents(mice, eventList, 'Eat_Start', trials)
eventEnd = getEvents(mice, eventList, 'Eat_End', trials)

# Plot the dFF data
plt.figure(figsize=(20,10), facecolor="w", dpi= 150)
for i, [mus, sess] in enumerate(mice):
    ax = plt.subplot(len(mice),1,i+1)
    dFFList[mus].plot(ax=ax, legend=False, cmap=plt.cm.rainbow)
    ax.set_ylabel("dF/F (%)")
    
    # Plot the eating bouts
    for i, event in enumerate(eventStart[mus]):
        ax.fill_between([eventStart[mus].loc[i+1], eventEnd[mus].loc[i+1]], -10, 50, facecolor='gray', linewidth=0, alpha=0.3, zorder=-55)
```

```python
# dFF and plot again
filtList = filtData(mice, dffCalc(mice, dataList, lowest=False), cutoff=5.0, order=6)
dFFList = smoothData(mice, filtList, window=4)

# Normalize the data
normList = normData(mice, dFFList)

# Get the events
eventStart = getEvents(mice, eventList, 'Eat_Start', trials)
eventEnd = getEvents(mice, eventList, 'Eat_End', trials)

if 0:
    # Plot the dFF data
    n = 100
    fig = plt.figure(figsize=(50,50), facecolor="w", dpi= 150)
    for i, [mus, sess] in enumerate(mice):    
        offset = np.linspace(0, n, len(dFFList[mus].columns))
        ax = plt.subplot(len(mice),1,i+1)
        (dFFList[mus]+offset).plot(ax=ax, legend=False, cmap=plt.cm.rainbow)
        ax.set_ylabel("dF/F (%)")

        # Plot the eating bouts
        for i, event in enumerate(eventStart[mus]):
            ax.fill_between([eventStart[mus].loc[i+1], eventEnd[mus].loc[i+1]], -20, 150, facecolor='gray', linewidth=0, alpha=0.3, zorder=-55)

    fig.savefig("J:\\Hakan Kucukdereli\\Miniscope_Behaviour_MPIN\\scripts\\figures\\first_draft\\1_all_cells_filt2_dff.svg")
```

```python
# Plotting the lines up events
# Pick the event tyoe anad event parameters
save = False
eventType = 'Eat_Start'
base = -10.0
duration = 30.0

# Find the marked events
eventsData = markTrials(mice, dFFList, base=base, duration=duration, eventType=eventType, behType=behType, trials=trials, baselining=True)
eventsData_norm = markTrials(mice, normList, base=base, duration=duration, eventType=eventType, behType=behType, trials=trials, baselining=True)

# Calculate the mean and sem
mean=eventsData.pivot_table(index=['Event', 'Cell'], columns='New_Time').mean()
sem=eventsData.pivot_table(index=['Event', 'Cell'], columns='New_Time').sem()
time = np.arange(base,duration,fs)

fig, ax = plotMean(mean, sem, time, eventType)
fname = "J:\\Hakan Kucukdereli\\Miniscope_Behaviour_MPIN\\scripts\\for Amelia\\"+"2_"+eventType+"_all_trials_mean_base_"+str(base)+"sec.svg"
if save:    
    fig.savefig(fname)
```

```python
# Plotting the heatmap from normalized data
normTrials=eventsData.pivot_table(index='Event', columns='New_Time', values='Fluoro')
mini = normTrials.min(axis=1)
maxi = normTrials.max(axis=1)

for i in np.arange(trials[0], trials[1]+1):
    normTrials.loc[i] = (normTrials.loc[i] - mini.loc[i]) / (maxi.loc[i] - mini.loc[i])
    
#plt.style.use('dark_background')
plt.style.use('classic')
# Set the font dictionaries (for plot title and axis titles)
font = {'sans-serif' : 'Arial',
        'weight' : 'normal',
        'size'   : 18,}
plt.rc('font', **font)

fig = plt.figure(figsize=(6,5), facecolor="w", dpi= 150)
ax = plt.subplot(111)

line = 1.5
colors = ('#BCBEC0', '#0070C0', '#BCBEC0', '#0070C0')

#plt.pcolor(normTrials, cmap=cm_r, linewidth=line, linestyle='solid', vmin=0.0, vmax=1.0)
plt.pcolor(normTrials, cmap=plt.cm.hot, linewidth=line, linestyle='solid')

ax.invert_yaxis()
cbar = plt.colorbar(ax=ax, ticks=np.arange(0.0, 1.001, 0.2))
cbar.set_label('$\Delta$F/F$_0$ (%)')

## labels
ax.set_xlabel('Time (sec)')
ax.set_ylabel('Trial #')
ax.set_title(eventType)

time_ax = np.arange(base, duration+0.001, 5.0)
#ax.set_xticks(np.linspace(0, (-base+duration)/fs, len(time_ax)))
#ax.set_xticklabels(time_ax)
ax.set_xticks(np.linspace(0, (-base+duration)/fs, 5))
ax.set_xticklabels([-10, 0, 10, 20, 30])
ax.set_yticks(np.linspace(trials[0]-0.5, trials[1]-0.5, 3))
ax.set_yticklabels(np.linspace(trials[0], trials[1], 3, dtype=int))

ax.plot([-base/fs, -base/fs], list(ax.get_ylim()), 'w', linestyle='--', linewidth=line*1.2, alpha=1., zorder=111)
fname = "J:\\Hakan Kucukdereli\\Miniscope_Behaviour_MPIN\\scripts\\figures\\first_draft\\heatmap_options\\new_heatmaps\\"+"dark_2_hot"+eventType+ "_all_trials_heatmap_"+str(base)+"sec.svg"
if save:
    fig.savefig(fname)
```
